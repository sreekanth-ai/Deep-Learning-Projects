```python
## Importing the image module from matplotlib library

import matplotlib.image as mpimg
import matplotlib.pyplot as plt
```

```python
## Loading an image through matplotlib.image module

img = mpimg.imread('/content/dog.jpg')
```

```python
## Image Dimension
img.shape
```

**It's a 3 matrix image, means RGB image and it has 3 channels.**

```python
## Image array
print(img)
```

**Displaying the image from numpy array**

```python
img_plot = plt.imshow(img)
plt.show()
```

**Resizing Image using Pillow library**

```python
from PIL import Image
```

```python
img = Image.open('/content/dog.jpg')

img_resized = img.resize((200,200))
```

```python
img_resized.save('dog_img_resized.jpg')
```

```python
img_res = mpimg.imread('/content/dog_img_resized.jpg')
img_res_plot = plt.imshow(img_res)
plt.show()
```

**Converting RGB images to Grayscale**

```python
## Importing openCV library
import cv2
```

```python
img = cv2.imread('/content/dog.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_img = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_img)
```

```python
grayscale_img.shape
```

**Grayscale images are 1 channel image**

*    cv2.imshow() will display the image. But this will not be allowed in google colab

**from google.colab.patches import cv2_imshow**

```python
from google.colab.patches import cv2_imshow
```

```python
## Displaying the image
cv2_imshow(grayscale_img)
```

```python
## Saving the grayscale image
cv2.imwrite('dog_grayscale_img.jpg',grayscale_img)
```
